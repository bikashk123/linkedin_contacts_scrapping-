```python
URL="https://www.flipkart.com/realme-xt-pearl-blue-64-gb/p/itm731360fdbd273?pid"\
     "=MOBFJYBE9FHXFEFJ&lid=LSTMOBFJYBE9FHXFEFJNVQVIV&marketplace=FLIPKART&srno=s_1_1&otracker=search&otracker1"\
   "=search&fm=SEARCH&iid=ccc89f62-296e-429b-a78e-f3d449bd9449.MOBFJYBE9FHXFEFJ"\
".SEARCH&ppt=sp&ppn=sp&ssid=2vj1vpafcw0000001600769011158&qH=3d8bbf69b4087ce1"
```

```python
req_data=requests.get(URL) #return data in string format
review_soup=BeautifulSoup(req_data.content,'html.parser') #do parsing
```

```python
#review_soup
```

```python
all_reviews=review_soup.find_all('div',{'class':'col _39LH-M'})
```

```python
all_reviews
```

```python
rating_list=[]
review_header_list=[]
detailed_review_list=[]
user_list=[]
likes_dislikes_list=[]

for review in all_reviews:
    rating=review.find_all('div',{'class':'hGSR34 E_uFuv'})
    text=[e.get_text() for e in rating]
    review_header=review.find_all('p',{'class':'_2xg6Ul'})
    review_header=[e.get_text() for e in review_header]
    
    detailed_review=review.find_all('div',{'class':'qwjRop'})
    detailed_review=[e.get_text() for e in detailed_review]
    user=review.find_all('p',{'class':'_3LYOAd _3sxSiS'})
    user=[e.get_text() for e in user]
    likes_dislikes=review.find_all('span',{'class':'_1_BQL8'})
    likes_dislikes=[e.get_text() for e in likes_dislikes]
    
    rating_list.append(text)
    review_header_list.append(review_header)
    detailed_review_list.append(detailed_review)
    user_list.append(user)
    likes_dislikes_list.append(likes_dislikes)
```

```python
print(review_header_list)
```

```python
data={'rating':rating_list,'review_header':review_header_list,'detailed_review':detailed_review_list,'user':user_list,'likes_dislikes':likes_dislikes_list}
```

```python
import pandas as pd
df = pd.DataFrame(data, columns=['rating','review_header','detailed_review','user','likes_dislikes'])
```
